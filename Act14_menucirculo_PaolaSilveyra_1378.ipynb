#### 📦 Arquitectura
Programación procedural y modular. Todo el código reside en un único archivo .py. No se utiliza programación orientada a objetos (clases ni herencia).
#### 🔄 Flujo de Control
Usa while True: para mantener el menú activo indefinidamente. if / elif / else dirige la ejecución según la opción elegida. break termina el ciclo cuando el usuario selecciona 4.
#### 🔗 Modularidad
4 funciones claramente separadas: areacirculo(), perimetrocirculo(), volumenesfera() y menu(). Cada función de cálculo se invoca directamente desde el menú.
#### 📐 Precisión Matemática
Importa math.pi para usar el valor real de π. Aplica las fórmulas correctas:
• Área: π·r²
• Perímetro: 2·π·r
• Volumen esfera: (4/3)·π·r³
#### 💡 Interactividad

```python
# ------------------------------------------------------
# PROGRAMA DE CALCULO GEOMETRICO 📐
# Elaborado por Paola Silveyra
# Formulas utilizadas:
# Área del círculo =  A=π⋅r2A=π⋅r 2
# Perímetro del círculo =  P=2⋅π⋅rP=2⋅π⋅r
# Volumen de la esfera =  V=43⋅π⋅r3V= 34​⋅π⋅r 3
# ------------------------------------------------------

import math

# ──────────────────────────────────────────────────────
# FUNCIÓN: Cálculo del área del círculo
# ──────────────────────────────────────────────────────
def areacirculo():
    print("\n🔵 Cálculo del Área del Círculo")
    radio = float(input("📏 Ingresa el radio: "))
    area = math.pi * radio ** 2
    print(f"✅ Resultado: El área es {area:.2f} u²\n")

# ──────────────────────────────────────────────────────
# FUNCIÓN: Cálculo del perímetro del círculo
# ──────────────────────────────────────────────────────
def perimetrocirculo():
    print("\n🟢 Cálculo del Perímetro del Círculo")
    radio = float(input("📏 Ingresa el radio: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ Resultado: El perímetro es {perimetro:.2f} u\n")

# ──────────────────────────────────────────────────────
# FUNCIÓN: Cálculo del volumen de la esfera
# ──────────────────────────────────────────────────────
def volumenesfera():
    print("\n🔴 Cálculo del Volumen de la Esfera")
    radio = float(input("📏 Ingresa el radio: "))
    volumen = (4 / 3) * math.pi * radio ** 3
    print(f"✅ Resultado: El volumen es {volumen:.2f} u³\n")

# ──────────────────────────────────────────────────────
# FUNCIÓN: Menú interactivo principal
# ──────────────────────────────────────────────────────
def menu():
    while True:
        print("\n" + "✨" * 25)
        print("   🧮 CALCULADORA GEOMÉTRICA 🧮")
        print("✨" * 25)
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")
        print("✨" * 25)

        opcion = input("👉 Selecciona una opción (1-4): ")

        if opcion == "1":
            areacirculo()               # Llamada a función
        elif opcion == "2":
            perimetrocirculo()          # Llamada a función
        elif opcion == "3":
            volumenesfera()             # Llamada a función
        elif opcion == "4":
            print("\n👋 ¡Gracias por usar la calculadora! Hasta pronto. 🚀\n")
            break                       # Sale del ciclo while
        else:
            print("\n⚠️  Opción no válida. Por favor, ingresa 1, 2, 3 o 4.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
