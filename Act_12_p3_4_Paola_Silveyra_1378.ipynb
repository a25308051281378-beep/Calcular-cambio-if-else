## las variables
1. CT--> Costo del traje
2. DE--> Descuento
3. PF--> Precio final

## Formulas
* $$DE = CT \times 0.08$$
* $$PF = CT - DE$$


## Diagrama de flujo lucidChart
<img src="https://raw.githubusercontent.com/a25308051281378-beep/Descuento-y-precio-final-p3.4-/refs/heads/main/DF%20traje%20p3%2C4.png" width="600" height="700" alt="Texto descriptivo">

```python
# para cuando es verdadero (si)
# calculando el descuento y precio final usando if else
# Datos de entrada: CT (costo del Traje)
# Si es verdadero (CT > 2500) aplicar 15%
# Si es falso (else) aplicar 8%

CT = float(input("Ingrese el costo del traje: "))

if CT > 2500:
  DE = CT * 0.15
  print("El descuento es el 15%: ", DE)
else:
  DE = CT * 0.08
  print("El descuento es del 8%: ", DE)

  PF = CT - DE
  print("El precio final es: ", PF)

  print("Fin del programa")
  print("Programa realizado por Paola Silveyra NL = 50")
```

```python
# para cuando es falso (no)
# Calculando el descuento y precio final usando if else
# Datos de entrada: CT (Costo del Traje)
# Si es verdadero (CT > 2500) aplicar 15%
# Si es falso (else) aplicar 8%

CT = float(input("Ingrese el costo del traje: "))

if CT > 2500:
  DE = CT * 0.15
  print("El descuento es del 15%: ", DE)
else:
    DE = CT * 0.08
    print("El descuento es del 8%: ", DE)

    PF = CT - DE
    print("El precio final es: ", PF)

    print("Fin del programa")
    print("Programa realizado por Paola Silveyra NL = 50")


```
